*   **Ladder score:** skor kebahagiaan atau kesejahteraan subjektif
*   **Logged GDP Per capita:** Rangkaian PDB per kapita dari tahun 2021menggunakan perkiraan pertumbuhan PDB riil di setiap negara
*   **Social support:** dukungan sosial mengacu pada bantuan atau dukungan yang diberikan oleh jaringan sosial kepada seseorang
*   **Healthy life expectancy:** angka harapan hidup sehat adalah rata-rata usia hidup dalam kondisi sehat
*   **Freedom to make life choice:** kebebasan untuk membuat pilihan hidup
*   **Generosity:** kedermawanan dari respons terhadap pertanyaan GWP “Apakah Anda menyumbangkan uang ke badan amal dalam sebulan terakhir?” pada PDB per kapita
*   **Perceptions of corruption:** Ukurannya adalah rata-rata nasional dari respons survei terhadap dua pertanyaan dalam GWP: “Apakah korupsi tersebar luas di seluruh pemerintahan atau tidak” dan “Apakah korupsi tersebar luas di dunia usaha atau tidak?”
*   **Ladder score in dystopia:** memiliki nilai yang sama dengan rata-rata nasional terendah di dunia.  Dystopia adalah negara imajiner yang memiliki penduduk paling tidak bahagia di dunia

```python
data_numeric = data.select_dtypes(include=['float64', 'int64'])
```

```python
sns.heatmap(data_numeric.corr(), annot = True, fmt = ".2f", linewidth = .7)
plt.title("Relationship Between Features ")
plt.show()
```

dari hasil diatas dapat disimpulkan bahwa ladder_score (happiness score) dengan GDP, sosial support, healthy life expectancy, freedom to make life choices memiliki korelasi yang tinggi. sedangkan generosity dan corruption terdapat korelasi yang rendah

## Ladder Score by GDP

```python
plt.rcParams['figure.figsize'] = (15,7)
plt.title('Plot between Ladder Score and GDP')
sns.scatterplot(x = data.ladder_score, y = data.logged_GDP_per_capita,hue = data.regional_indicator, s = 200);

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Ladder score')
plt.ylabel('Logged GDP per capita')
```

Pada grafik di atas dapat disimpulkan bahwa negara dari Eropa Barat memiliki skor kebahagiaan tertinggi dan PDB per kapita tertinggi. Sedangkan Asia Selatan memiliki skor kebahagiaan terendah menurut data laporan kebahagiaan dunia tahun 2021.

```python
#GDP by Region
gdp_region = data.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
print(gdp_region)
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

Pada diagram lingkaran di atas menyatakan bahwa Afrika Sub-Sahara dan Eropa Barat memberikan kontribusi PDB tertinggi, sedangkan Amerika Utara dan ANZ memberikan kontribusi yang lebih kecil terhadap PDB dunia.

# Ladder Score by Healthy Life Expectancy

```python
#10 Negara dengan angka harapan hidup Tinggi dan Rendah
top_10 = data.head(10)
bottom_10 = data.tail(10)
# ha is horizontal allignment
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,6))
plt.tight_layout(pad = 2)
xlabels = top_10.country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels , rotation=45, ha = 'right')
sns.barplot(x = top_10.country_name, y = top_10.healthy_life_expectancy, ax = axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')


xlabels = bottom_10.country_name
axes[1].set_title('Bottom 10 happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels , rotation=45, ha = 'right')
sns.barplot(x = bottom_10.country_name, y = bottom_10.healthy_life_expectancy, ax = axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')
```

Jadi dari barplot di atas kita dapat dengan jelas memvisualisasikan bahwa negara-negara yang paling bahagia memiliki angka harapan hidup di atas 70 tahun, sementara 10 negara yang paling tidak bahagia memiliki angka harapan hidup kurang dari 60 tahun.
